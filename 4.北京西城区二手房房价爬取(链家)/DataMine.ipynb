```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
plt.style.use('ggplot')
# 用于正常显示中文
plt.rcParams['font.family'] = ['Heiti']
```

```python
price=pd.read_csv('西城区二手房房价信息清洗1.csv')
```

```python
price.head()
```

```python
price.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
price.head()
```

```python
sns.countplot('所在区域',data=price,)
```

```python
maxV=price['单价(元/平米)'].max()
```

```python
price[price['单价(元/平米)']==maxV]
```

```python
price[price['单价(元/平米)']==price['单价(元/平米)'].min()]
```

```python
averageV=price['单价(元/平米)'].sum()/3000
```

```python
averageV
```

```python
plt.figure()
sns.countplot('位置名称',data=price)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='总价(万)',y='单价(元/平米)',data=price)
#plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='位置名称',y='单价(元/平米)',data=price)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='位置名称',y='总价(万)',data=price)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure()
sns.distplot(price['单价(元/平米)'])
plt.xlim(50000,180000)
#plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure()
sns.distplot(price['总价(万)'],bins=100)
#plt.xticks(rotation='vertical')
plt.xlim(200,1000)
plt.show()
```
