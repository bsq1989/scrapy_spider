```python
import pandas as pd
```

```python
price=pd.read_csv('housePrice_bj_xicheng.csv')
```

```python
price.head()
```

```python
price.drop('index',axis=1,inplace=True)
```

```python
price.head()
```

```python
price['所在区域']
```

```python
price['所在区域'].apply(lambda x:x.strip())
```

```python
def stripStr(x):
    return x.strip()
```

```python
price['所在区域']=price['所在区域'].apply(stripStr)
```

```python
price.head()
```

```python
def removeXiCheng(x):
    return x[3:]
```

```python
price['所在区域']=price['所在区域'].apply(removeXiCheng)
```

```python
price['位置名称']=price['所在区域'].apply(lambda x:x.split()[0])
```

```python
def removePosName(x):
    parts=x.split()
    if len(parts)>1:
        return parts[1]
    else:
        return ' '
```

```python
price['所在区域']=price['所在区域'].apply(removePosName)
```

```python
price.head()
```

```python
price['总价(万)']=price['总价']
```

```python
price.drop('总价',axis=1,inplace=True)
```

```python
price['单价(元/平米)']=price['单价']
```

```python
price.drop('单价',axis=1,inplace=True)
```

```python
price.drop('总价单位',axis=1,inplace=True)
```

```python
price.drop('单价单位',axis=1,inplace=True)
```

```python
price.head()
```

```python
temp=price['房屋年代'].apply(lambda x:'/' in x)
```

```python
def getBuildType(x):
    return x.split('/')[1]

price['建筑类型']=price['房屋年代'].apply(getBuildType)
```

```python
import re
```

```python
def getDynasty(x):
    pattern=re.compile('\d*')
    return pattern.findall(x)[0]
price['房屋年代']=price['房屋年代'].apply(getDynasty)
```

```python
price.head()
```

```python
temp=price['楼层'].apply(lambda x:'(' in x)
```

```python
def getTotalFloors(x):
    temp=x.find('共')
    return x[temp+1:]
def getFloor(x):
    temp=x.find('(')
    return x[:temp]
```

```python
price['楼层总数']=price['楼层'].apply(getTotalFloors)
```

```python
price['楼层']=price['楼层'].apply(getFloor)
```

```python
price['楼层总数']=price['楼层总数'].apply(getDynasty)
```
